```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
# Using the top part as a header, change that 
# using white space as the delimiter

dframe = pd.read_csv('Sample.txt', delim_whitespace=True)
dframe.head()
```

```python
dframe = pd.read_csv('Sample.txt', delim_whitespace=True, header=None)
dframe.head()
```

```python
# Another way to add a dataframe
dframe = pd.read_table('Sample.txt', delim_whitespace=True, header=None)
dframe.head()
```

```python
# CSV file, only want the first two files

pd.read_csv('Sample.csv', header=None, delim_whitespace=True, nrows=2)
```

```python
# Changing the file towards the csv file
# The first parameter is the file you want to call
dframe.to_csv('new_sample.csv')
```
