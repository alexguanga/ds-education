```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
dframe1 = DataFrame({'key':['X','Z','Y', 'Z','X','X'], 'data_set_1': np.arange(6)})
dframe1
```

```python
dframe2 = DataFrame({'key':['Q','Y','Z'], 'data_set_2': np.arange(1,4,1)})
dframe2
```

```python
pd.merge(dframe1, dframe2)
```

```python
# specifying the way I want them to link
pd.merge(dframe1, dframe2, on='key')
```

```python
# everything from dframe1 is passed
pd.merge(dframe1, dframe2, on='key', how='left')
```

```python
pd.merge(dframe2, dframe1, on='key', how='left')
```

```python
# Union
pd.merge(dframe2, dframe1, on='key', how='outer')
```

```python
dframe3 = DataFrame({'key':['X','X','X', 'Y','Z','Z'], 'data_set_3': np.arange(6)})
dframe3
```

```python
dframe4 = DataFrame({'key':['Y','Y','X', 'X','Z'], 'data_set_4': np.arange(5)})
dframe4
```

```python
# merges every values from one dataframe to the other dataframe
pd.merge(dframe3, dframe4)
```

```python
df_left = DataFrame({'key':['SF','SF','LA'], 
                     'key2': ['one','two','one'], 
                     'left_data': [10,20,3]})
df_left
```

```python
df_right = DataFrame({'key':['SF','SF','LA', 'LA'], 
                     'key2': ['one','two','one', 'two'], 
                     'right_data': [40,50,60,70]})
df_right
```

```python
pd.merge(df_left, df_right, on=['key', 'key2'], how='outer')
```

```python
# Checking the suffix value, does it by itself
pd.merge(df_left, df_right, on='key')
```

```python
# creating your own suffix
pd.merge(df_left, df_right, on='key', suffixes=('_lefty', '_righty'))
```
